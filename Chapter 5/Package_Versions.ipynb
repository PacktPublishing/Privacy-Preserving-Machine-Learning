# List of pakcages and versions

```python
import pkg_resources

installed_packages = pkg_resources.working_set

installed_packages_list = sorted(["%s==%s" % (i.key, i.version) for i in installed_packages])

for i in range ( 1, len(installed_packages_list)):
    print(installed_packages_list[i])
    
```

```python
import csv
with open('Installed Packages.csv', 'w') as f:
    csv_writer = csv.writer(f)
    csv_writer.writerow(installed_packages_list)
```
